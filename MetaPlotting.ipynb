```julia
linspace(a,b,l)=range(a,stop=b,length=l)
x=range(-10,stop=10,length=100)
y=range(-10,stop=10,length=100)
z(x,y) = sin(x)+cos(y)
```

```julia
surface(x,y,z)
```

```julia
x = range(2,stop=40,length=200)
surface(x,x,z)
```

```julia
gr()
x = linspace(-5,5,50)
y = linspace(-3,10,30)
ff(x,y) = sin(x) + x*cos(y)^2
sts = [:contour, :heatmap, :surface, :wireframe]
plot(x, y, ff, st = sts, layout = length(sts))
```

```julia
gr(reuse=true);@gif for i in linspace(0,2π,50)
    X = Y = linspace(-5,5,40)
    surface(X, Y, (x,y) -> sin(x+10cos(i))+cos(y))
end
```

```julia
using Plots
pyplot(leg=false, ticks=nothing)
x = y = linspace(-5, 5, 40)
zs = zeros(0,40)

@gif for i in linspace(0, 2π, 100)
    f(x,y) = sin(x + 10sin(i)) + cos(y)

    # create a plot with 3 subplots and a custom layout
    l = @layout [a{0.7w} b; c{0.2h}]
    p = plot(x, y, f, st = [:surface, :contourf], layout=l)

    # add a tracking line
    fixed_x = zeros(40)
    zz = map(f,fixed_x,y)
    plot!(p[1], fixed_x, y, zz, line = (:black, 5, 0.2))
    vline!(p[2], [0], line = (:black, 5))

    # add to and show the tracked values over time
    global zs = vcat(zs, zz')
    plot!(p[3], zs, alpha = 0.2, palette = cgrad(:blues).colors)
end
```
